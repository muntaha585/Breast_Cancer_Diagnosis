# Radial Basis Function Network (RBFN): For anomaly detection

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

class RBFN:
    def __init__(self, n_centers, learning_rate, max_iters):
        self.n_centers = n_centers
        self.learning_rate = learning_rate
        self.max_iters = max_iters
        self.centers = None
        self.weights = None
        self.sigma = None

    def _kernel_function(self, x, center):
        """Gaussian RBF kernel."""
        return np.exp(-np.linalg.norm(x - center)**2 / (2 * self.sigma**2))

    def fit(self, X, y):
        # 1. Determine RBF centers (e.g., using K-Means)
        kmeans = KMeans(n_clusters=self.n_centers, random_state=42, n_init=10)
        kmeans.fit(X)
        self.centers = kmeans.cluster_centers_

        # 2. Determine the RBF kernel width (sigma)

        max_dist = 0
        for i in range(self.n_centers):
            for j in range(i + 1, self.n_centers):
                dist = np.linalg.norm(self.centers[i] - self.centers[j])
                if dist > max_dist:
                    max_dist = dist
        self.sigma = max_dist / np.sqrt(2 * self.n_centers) if self.n_centers > 1 else 1.0


        # 3. Build the RBF activation matrix
        G = np.zeros((X.shape[0], self.n_centers))
        for i in range(X.shape[0]):
            for j in range(self.n_centers):
                G[i, j] = self._kernel_function(X[i], self.centers[j])

        # Add a bias term to the G matrix
        G = np.concatenate((np.ones((G.shape[0], 1)), G), axis=1)


        # 4. Learn the output layer weights using pseudo-inverse (closed-form solution)
        self.weights = np.linalg.pinv(G).dot(y)

    def predict(self, X):
        # Build the RBF activation matrix for the input data
        G = np.zeros((X.shape[0], self.n_centers))
        for i in range(X.shape[0]):
            for j in range(self.n_centers):
                G[i, j] = self._kernel_function(X[i], self.centers[j])

        # Add bias term
        G = np.concatenate((np.ones((G.shape[0], 1)), G), axis=1)

        # Compute the output
        output = G.dot(self.weights)

        return output

# Initialize and train the RBFN
n_rbf_centers = 50
rbfn = RBFN(n_centers=n_rbf_centers, learning_rate=0.01, max_iters=100)

# Train the RBFN on the training data
print("\n--- Training Radial Basis Function Network (RBFN) ---")
rbfn.fit(X_train, y_train)
print("--- RBFN Training Complete ---")

# Predict the output for the test data
y_pred_rbfn_output = rbfn.predict(X_test)

plt.figure(figsize=(10, 6))
sns.histplot(y_pred_rbfn_output, kde=True)
plt.title('Distribution of RBFN Output on Test Data')
plt.xlabel('RBFN Output Value')
plt.ylabel('Frequency')
plt.show()

output_benign = y_pred_rbfn_output[y_test == 0]
output_malignant = y_pred_rbfn_output[y_test == 1]

plt.figure(figsize=(12, 6))
sns.histplot(output_benign, kde=True, color='blue', label='True Benign (0)')
sns.histplot(output_malignant, kde=True, color='red', label='True Malignant (1)')
plt.title('RBFN Output Distribution by True Class')
plt.xlabel('RBFN Output Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

anomaly_scores = np.abs(y_pred_rbfn_output - y_test)

plt.figure(figsize=(10, 6))
sns.histplot(anomaly_scores, kde=True)
plt.title('Distribution of RBFN Anomaly Scores (Absolute Error)')
plt.xlabel('Anomaly Score')
plt.ylabel('Frequency')
plt.show()

anomaly_threshold = np.quantile(anomaly_scores, threshold_quantile)

potential_anomalies_indices = np.where(anomaly_scores > anomaly_threshold)[0]
potential_anomalies = X_test[potential_anomalies_indices]
potential_anomalies_true_labels = y_test[potential_anomalies_indices]
potential_anomalies_predicted_output = y_pred_rbfn_output[potential_anomalies_indices]

print(f"\n--- Potential Anomalies detected with RBFN (Anomaly Score > {anomaly_threshold:.4f}) ---")
print(f"Number of potential anomalies detected: {len(potential_anomalies_indices)}")
print("\nDetails of potential anomalies:")
for i, idx in enumerate(potential_anomalies_indices):
    print(f"Sample Index in Test Set: {idx}")
    print(f"  True Label: {label_encoder.inverse_transform([potential_anomalies_true_labels[i]])[0]}")
    print(f"  RBFN Output: {potential_anomalies_predicted_output[i]:.4f}")
    print(f"  Anomaly Score: {anomaly_scores[idx]:.4f}")
    print("-" * 20)

if X_test.shape[1] >= 2:
    plt.figure(figsize=(10, 8))
    # Plot all test points colored by true label
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='viridis', alpha=0.5, label='All Test Points')
    # Highlight potential anomalies
    plt.scatter(potential_anomalies[:, 0], potential_anomalies[:, 1], c='red', marker='X', s=100, label=f'Potential Anomalies (Score > {anomaly_threshold:.4f})')
    plt.title('RBFN Anomaly Detection Visualization (using first 2 features)')
    plt.xlabel(df.columns[0])
    plt.legend()
    plt.colorbar(label='True Diagnosis (0: Benign, 1: Malignant)')
    plt.grid(True)
    plt.show()
else:
    print("\nCannot plot anomaly visualization on 2D scatter plot: dataset has fewer than 2 features.")

```

# Bidirectional Associative Memory (BAM): For associating symptoms with diseases

```python

bam_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst']

df_bam = df[bam_features + ['diagnosis']].copy()


df_train_bam_features = pd.DataFrame(X_train, columns=X.columns)[bam_features]
feature_medians = df_train_bam_features.median()

# Apply binary discretization to the selected features (using the training medians)
for feature in bam_features:
    # Use .loc to avoid SettingWithCopyWarning
    df_bam.loc[:, feature] = (df_bam[feature] >= feature_medians[feature]).astype(int)

# The diagnosis is already 0 or 1, but let's ensure it's integer type
df_bam['diagnosis'] = df_bam['diagnosis'].astype(int)

# Display the binary data for BAM
print("\n--- Binary Data for BAM ---")
print(df_bam.head())

bam_data_features = df_bam[bam_features].values
bam_data_diagnosis = df_bam['diagnosis'].values.reshape(-1, 1) # Reshape to a column vector

X_bipolar = 2 * bam_data_features - 1
Y_bipolar = 2 * bam_data_diagnosis - 1

W = np.zeros((Y_bipolar.shape[1], X_bipolar.shape[1]))

for i in range(X_bipolar.shape[0]):
    W += np.outer(Y_bipolar[i], X_bipolar[i])

print("\n--- BAM Association Matrix (W) ---")
print(W)

X_train_bipolar = 2 * (df_bam[bam_features].values) - 1
Y_train_bipolar = 2 * (df_bam['diagnosis'].values.reshape(-1, 1)) - 1


Y_recall_bipolar = np.sign(X_train_bipolar @ W.T)


Y_recall_binary = (Y_recall_bipolar + 1) / 2

original_diagnosis_binary = df_bam['diagnosis'].values
recall_accuracy = np.mean(Y_recall_binary.flatten() == original_diagnosis_binary)

print(f"\n--- BAM Recall Test on Training Data (Features -> Diagnosis) ---")
print(f"Accuracy of recalling Diagnosis from Features: {recall_accuracy:.4f}")


plt.figure(figsize=(10, 2))
sns.heatmap(W, annot=True, cmap='coolwarm', center=0)
plt.title('BAM Association Matrix (Features vs. Diagnosis)')
plt.xlabel('Feature (Binary)')
plt.ylabel('Diagnosis (Binary)')
plt.yticks([0.5], ['Diagnosis'], rotation=0)
plt.xticks(np.arange(len(bam_features)) + 0.5, bam_features, rotation=45, ha='right')
plt.tight_layout()
plt.show()


if X_test.shape[1] >= len(bam_features):

    X_test_original_scale = scaler.inverse_transform(X_test)[:, [df.columns.get_loc(col) for col in bam_features if col in df.columns]]

    # Create a temporary DataFrame for easier median comparison
    df_test_bam_features = pd.DataFrame(X_test_original_scale, columns=bam_features)

    # Discretize the test features
    test_pattern_binary = np.zeros((X_test_original_scale.shape[0], len(bam_features)))
    for i, feature in enumerate(bam_features):
        test_pattern_binary[:, i] = (df_test_bam_features.iloc[:, i] >= feature_medians[feature]).astype(int)

    # Select the first test sample's binary features and convert to bipolar
    single_test_pattern_binary = test_pattern_binary[0].reshape(1, -1)
    single_test_pattern_bipolar = 2 * single_test_pattern_binary - 1

    # Perform BAM recall for this test pattern
    recalled_diagnosis_bipolar = np.sign(single_test_pattern_bipolar @ W.T)
    recalled_diagnosis_binary = (recalled_diagnosis_bipolar + 1) / 2

    print("\n--- BAM Recall for the First Test Sample ---")
    print(f"Test Sample Binary Features: {single_test_pattern_binary.flatten()}")
    print(f"Recalled Diagnosis (Binary): {int(recalled_diagnosis_binary[0, 0])}")
    print(f"True Diagnosis (Binary): {y_test[0]}")
    print(f"True Diagnosis (Original): {label_encoder.inverse_transform([y_test[0]])[0]}")


    # Evaluate BAM performance on the binary test set (using the same median thresholds)
    Y_test_binary = y_test.reshape(-1, 1)
    X_test_bipolar_for_bam = 2 * test_pattern_binary - 1
    Y_test_bipolar = 2 * Y_test_binary - 1

    Y_pred_bam_bipolar = np.sign(X_test_bipolar_for_bam @ W.T)
    Y_pred_bam_binary = (Y_pred_bam_bipolar + 1) / 2

    bam_test_accuracy = accuracy_score(Y_test_binary, Y_pred_bam_binary)

    print(f"\n--- BAM Performance on Binary Test Data ---")
    print(f"Test Accuracy (BAM): {bam_test_accuracy:.4f}")
    print("Classification Report (BAM):")
    print(classification_report(Y_test_binary, Y_pred_bam_binary))
    print("Confusion Matrix (BAM):")
    print(confusion_matrix(Y_test_binary, Y_pred_bam_binary))

    sns.heatmap(confusion_matrix(Y_test_binary, Y_pred_bam_binary), annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix (BAM)')
    plt.xlabel('Predicted Diagnosis (Binary)')
    plt.ylabel('True Diagnosis (Binary)')
    plt.show()

else:
    print("\nSkipping BAM evaluation: Not enough features available for the selected BAM features.")

```

```python
symptoms = df.drop('diagnosis', axis=1).columns.tolist()
diseases = df['diagnosis'].unique()

bam_matrix = np.zeros((len(symptoms), len(diseases)))

for i in range(len(diseases)):
  # This creates some dummy rules
  bam_matrix[i % len(symptoms),i] = 1
  bam_matrix[(i+1) % len(symptoms),i] = 1
  bam_matrix[(i+2) % len(symptoms),i] = 1

def diagnose(patient_symptoms):
  """
  Diagnoses a patient based on their symptoms using the BAM.
  Args:
      patient_symptoms: A list of symptoms (strings).
  Returns:
      A list of possible diseases (strings), ordered by association strength.
  """

  symptom_vector = np.zeros(len(symptoms))
  for symptom in patient_symptoms:
      if symptom in symptoms:
          symptom_vector[symptoms.index(symptom)] = 1
      else:
          print(f"Warning: Unknown symptom '{symptom}'")

  disease_vector = np.dot(bam_matrix.T, symptom_vector)
  # Convert to disease names based on vector indices
  possible_diseases = []
  for i in range(len(diseases)):
    possible_diseases.append((diseases[i], disease_vector[i]))
  possible_diseases.sort(key=lambda x: x[1], reverse=True)

  return possible_diseases

# Example usage:
patient1_symptoms = ['radius_mean', 'texture_mean']
diagnosis_result = diagnose(patient1_symptoms)

# Print diagnosis results
print(f"Diagnosis for patient with symptoms {patient1_symptoms}")
for disease, score in diagnosis_result:
    print(f"Possible Disease: {disease}, Association Score: {score}")

```
